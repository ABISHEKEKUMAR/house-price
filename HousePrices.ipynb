```python
for i in df:
    if df[i].dtype==object:
        x=df[i].mode()[0]
        df[i].fillna(x,inplace=True)
    else:
        x=df[i].mean()
        df[i].fillna(x,inplace=True)
    
```

```python
df.isnull().sum()
```

```python
for i in df:
    if df[i].dtype==object:
        from sklearn.preprocessing import LabelEncoder
        le=LabelEncoder()
        df[i]=le.fit_transform(df[i])
```
